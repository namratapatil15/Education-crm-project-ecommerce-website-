```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from fer import FER
from PIL import Image
```

```python
detector = FER(mtcnn=True)
```

```python
from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    image_path = filename
    break  # Only first image

# Load the image
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.title("Uploaded Image")
plt.show()
```

```python
# Detect emotions in the image
results = detector.detect_emotions(image_rgb)
```

```python
# Draw bounding boxes and labels
for result in results:
    (x, y, w, h) = result["box"]
    emotions = result["emotions"]
    dominant_emotion = max(emotions, key=emotions.get)

    cv2.rectangle(image_rgb, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(image_rgb, f'{dominant_emotion}', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.9, (255, 0, 0), 2)

# Display result
plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Emotion Detection Result")
plt.show()
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = detector.detect_emotions(frame_rgb)

    for result in results:
        (x, y, w, h) = result["box"]
        emotions = result["emotions"]
        dominant_emotion = max(emotions, key=emotions.get)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(frame, f'{dominant_emotion}', (x, y - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    cv2.imshow('Emotion Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
